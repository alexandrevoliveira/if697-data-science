```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
covid = pd.read_csv(r'caso.csv', encoding = "utf-8")
```

```python
covidRecife = covid[covid.city == 'Recife'].sort_values(by=['date'], ascending=True)
covidRecife.reset_index(inplace=True, drop=True)
covidRecife.tail()
```

```python
#quantidade de mortes total em relação ao dia
plt.plot(covidRecife['order_for_place'], (covidRecife['deaths']))
```

```python
#first try with a for loop

for index, i in enumerate(covidRecife.deaths):
    if index > 0:
        covidRecife['death_per_day'][index] = (covidRecife.deaths[index] - covidRecife.deaths[index-1])
         
covidRecife.tail()
```

```python
covidRecife['death_per_day'].plot.bar(figsize=(20,15))
```
